```python
# importing the library
import pandas as pd
```

```python
# creating a dictionary to load it as a data frame
peoples = {
    "first_name" : ["Corey", "Jane", "John"],
    "last_name" : ["Schafer", "Doe", "Doe"],
    "email" : ["CoreySchafer@mail.com", "JaneDoe@mail.com", "JohnDoe@mail.com"]
}
```

```python
# creating the data frame
peoples_df = pd.DataFrame(peoples)
```

```python
peoples_df
```

În cadrul acestui tutorial o să învățăm despre `Indexes`. Până acuma am văzut index default în partea a doua a tutorialului, dar acuma o să vedem cum putem seta indedex customatizat și care sunt beneficiile pentru a face asta.

Avem din nou un simplu data frame și o să începem partea de Index cu acesta după care o să punem în practică ce am învățat cu data frame-ul mai mare. Acest data frame are 3 coloane (first_name, last_name și email) și 3 rânduri. După cum se poate observa, în partea din stânga arată ca și cum ar fi o coloană, dar fără a avea un nume, acela este index-ul pentru acest data frame. Acest index default reprezintă un simplu interval de valori integer. Există anumite situații și cazuri când poate dorim să avem un identificator (index) diferit pentru fiecare rând. Acel indentifcator reprezintă label-ul pentru rândul respectiv, iar de cele mai multe ori este unic. Librăria Pandas nu ne obligă ca indexii să fie unici și există unele cazuri când nu o să fie, dar de cele mai multe ori o să fie unicate.

Pentru data frame-ul cu care lucrăm acuma, un index mai potrivit ar putea fi adresa de email, adresă care trebuie să fie unicată. Pentru a seta o coloană ca fiind un index o să ne folosim de metoa `set_index()`. Metoda respectivă se apelează pentru data frame-ul unde dorim să facem modificările, iar ca și argument se va trece sub formă de string numele coloanei ce dorim să o setăm ca index

```python
peoples_df.set_index('email')
```

Din moment ce am aplicat codul 'peoples_df.set_index('email') acum putem vedea că această coloană este în partea stângă a data frame-ului, iar datele din interiorul coloanei sunt boldate. Acest index pare ca și cum ar fi o coloană deoarece are un nume, are numele coloanei pe care am setat-o să fie index-ul pentru acest data frame. Putem să setăm acest index în acest mod, dar dacă afișăm din nou acest data frame poate o să avem parte de o surpriză.

```python
peoples_df
```

Putem observa că data frame-ul nostru nu s-a modificat, coloana de email nu a fost salvată ca fiind index-ul pentru acest data frame. Acest lucru se întâmplă deoarece librăria 'pandas' nu realizează aceste modificări 'in place' doar dacă îi specificăm noi acest lucru. Acest comportament ne permite să experimentăm fără a ne îngrijora că data frame-ul o să fie modifcat în anumite moduri. Să presupunem că dorim ca acel index să fie setat permanent, pentru asta la metoda 'set_inde()' o să îi mai oferim un argument care poartă denmirea `inplace` la care o să îi atribuim valoare 'True'

```python
peoples_df.set_index('email', inplace=True)
```

```python
peoples_df
```

În urma adăugării acelui argument acuma se poate oberva că s-a setat permanent acel index în cadrul data frame-ului. Putem să obține informații despre acel index utilizând argumentul `.index`

```python
peoples_df.index
```

Acel atribut ne returnează o listă cu toate valorile din cadrul index-ului și ne spune și care este numele acelui index (name='email')

De ce anume este așa ceva folositor? Adresa de emaill fiind un index ne oferă un identificator unic pentru fiecare rând, iar în partea a doua a tutorialului am utilizat indexatorul 'loc' pentru a căuta valori în data frame după label-uri, iar acest index reprezintă label-ul pentru rânduri

Înainte pentru indexatorul 'loc' utilizam valoarea 0 pentru a extrage primul rând din data frame deoarece acela era index-ul pentru rândul respectiv ('peoples_df.loc[0]'). Din moment ce am modificat index-ul să fie adresa de email (colana de email) putem să specificăm valoarea din acea coloană de pe un anumit rând pentru un indxator.

```python
peoples_df.loc['CoreySchafer@mail.com']
```

În cazul de mai sus am primit rândul pentru acel email ce a fost specificat în cadrul indexatorului 'loc'. Precum am văzut în cadrul părții precedente, putem să specificăm valori și pentru ce coloane dorim să extragem informații din acest data frame

```python
peoples_df.loc['CoreySchafer@mail.com', 'last_name']
```

În acest moment nu avem acele valori integer ca și index deoarece acuma utilizează adresa de email. Dacă dorim să folosim acele valori integer cum am folosit înainte o să primim o eroare. Primim eraorea deoarece nu mai are ca și index o valoare integer. Dacă dorim să utilzăm valori integer, atunci putem utiliza indexatorul 'iloc'

```python
peoples_df.loc[0]
```

```python
peoples_df.iloc[0]
```

În cazul în care am setat un index din greșeală și dorim să îl resetăm cum era la început, putem face asta utilizând metoda `reset_index()`. Acestei metode trebuie să îi atribuim doar argumentul 'inplace=True' dacă vrem să facem acele modifcări permanente 

```python
peoples_df.reset_index(inplace=True) 
```

```python
peoples_df
```

După ce am resetat index-ul acuma avem coloana 'email' ca și o coloană normală, iar în partea stângă a apărut din nou index-ul cum era înainte, cu numere integer. Dacă se știe de la început care coloană dorim sî fie index-ul de la bun început, putem să o stabilim atunci când creem data frame-ul în loc să o setăm mai târziu. Acest lucru se poate face când citim date dintr-un csv sau din alte surse.

O să trecem acuma la datele de la studiu pentru a vedea un exemplu concret de ce ne ajută partea de index.

```python
# loading the data
df = pd.read_csv('../data/survey_results_public.csv')
schema_df = pd.read_csv('../data/survey_results_schema.csv')
```

```python
# changing the max display setting for rows and columns
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
# printing the first 5 rows of the data frame
df.head()
```

În cadrul data frame-ului se observă la fel precum în data frame-ul cu care am lucrat anterior că și acesta are un anumit interval de numere integer pe partea stângă care nu au un nume de coloană. Acest interval reprezintă index-ul. Dacă ne uităm atent la acest data frame putem observa faptul că acest data frame pare să aibă o coloană care reține câte o valoare unică pentru fiecare persoană ce a răspuns, și anume coloana 'Respondent'. Am putea utiliza acea coloană 'Respondent' ca și index-ul principal pentru acest data frame

```python
df.set_index('Respondent', inplace=True)
```

```python
df.head()
```

Putem să facem cum am făcut mai sus, să utilizăm metoda 'set_index()' cu argumentele pentru numele coloanei și 'inplace=True'. O altă metodă de a face acest lucru, de a seta un index, este atunci când citim datele din fișierul csv. Metodei 'read_csv()' putem să îi mai oferim încă un argument , `index_col` prin care să îi precizăm o coloană care dorim să fie setată ca și index

```python
df = pd.read_csv('../data/survey_results_public.csv', index_col='Respondent')
```

```python
df.head()
```

Înainte de a seta index-ul ca fiind coloana 'Respondent' aveam în partea stângî 2 coloane care aveau amândouă valori de tip integer care practic rețineau același lucru. Din moment ce am setat coloana 'Respondent' ca și index, acuma pare mai aranjat un pic acest fata frame. Dacă dorim să extragem anumite informații din acest data frame, pentru indexatorul 'loc' putem să ne folosim acum de această coloană ce poartă statul de index. Valorile care sunt prezente în cadrul coloanei sunt de tip integer, prin urmare și la indexatorul 'loc' o să se treacă valori de tip integer

```python
df.loc[1]
```

Să ne uităm la un exemplu concret unde am utiliza această parte de index. În cadrul fișierului 'survey_results_schema' avem datele cu privire la coloane, și anume ce reprezintă fiecare coloană. Dacă dorim să vedem ce reprezintă coloana 'Hobbyist' trebuie să ne uităm în cadrul data frame-ului pentru schema acestui studiu și să căutăm valoare 'Hobbyist' în coloana denumită 'Column'

```python
schema_df
```

Întrebare din studiu pentru coloana 'Hobbyist' era 'Do you code as a hobby?'. Cum putem proceda în situația în care dorim să aflăm ce înseamnă o anumită coloană fără a fi nevoiți să căutăm prin tot data frame-ul de schema (deoarece are 85 de rânduri acest data frame). În cazul de față putem să setăm numele coloanei ca index pentru acest data frame, iar după să folosim indexatorul 'loc'

O să recitim datele din fișier și o să setăm index-ul în momentul citirii.

```python
schema_df = pd.read_csv('../data/survey_results_schema.csv', index_col='Column')
```

```python
schema_df
```

În acest moment putem vedea că datele din cadrul coloanei 'Column' sunt boldate acuma, prin urmare acestea reprezintă index-ul pentru acest data frame. Putem să ne folosim de indexatorul 'loc' și de aceste label-uri pentru a căuta direct aceste valori. Dacă dorim din nou să vedem ce anume reprezintă coloana 'Hobbyist' putem să utilizăm următorul cod:

```python
schema_df.loc['Hobbyist']
```

O îmbunătățire pentru a citi acest 'schema_df' data frame  ar fi dacă indexii ar fi sortați în ordine alfabetică (căutarea în data frame ar fi mult mai ușoară). Pentru a sorta un index o să ne folosim de metoda `sort_index()` 

```python
schema_df.sort_index()
```

Dacă se dorește ca sortarea să se facă descendent, atunci o să se adauge argumentul de `ascending=False`

```python
schema_df.sort_index(ascending=False)
```

Știm că aceste schimbări au loc doar pentru a vedea cum anume ar arăta modificările în data frame dacă acestea chiar s-ar produce. Pentru a face o modificare permanentă trebuie să utilizăm și aici argumentul 'inplace=True'

```python
schema_df.sort_index(inplace=True)
```

Din moment ce am executat codul de mai sus și nu am primit niciun output, asta înseamnă că modificările au fost făcute în cadrul data frame-ului. Dacă afișăm acuma data frame-ul, acum schimbările sunt definitive, index-ul acestui data frame este setat în ordine alfabetică.
