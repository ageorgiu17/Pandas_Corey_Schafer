O să aruncăm o privire peste un data frame și o să învățăm despre cum putem să ne imaginăm un data frame ca și un obiect în Python. În jupyter notebook, un data frame reprezintă datele sub formă de rânduri și coloane, practic un tabel.

Ceea ce a fost oferit ca și output pentru comanda 'df.head()' reprezintă un data frame. Un data frame este alcătuit din mai multe rânduri și mai multe coloane. Acest fișier conținând date despre un studiu, fiecare rând reprezintă răspunsurile unei persoane pentru fiecare întrebare din studiu respectiv. De exemplu, prima persoană din studiu a răspuns cu 'Yes' pentru coloana 'Hobbyist'. Dacă căutăm în cadrul fișierului 'survey_results_schema.csv' putem vedea că pentru coloana 'Hobbyist' întrebarea a fost 'Do you code as a hobby?'

Acum o să vedem cum anume ne gândim la un data frame utilizând doar limbaj de Python (fără anumite librării). Dacă ne folosim de Python doar pentru a stoca date sub formă de tipul rânduri și coloane, cum anume am face asta? Primul gând ar fi să stocăm informațiile respective în dicționarii. Să exerimentăm utilizând dicționariile.

```python
peoples = {
    "first_name" : "Corey",
    "last_name" : "Schafer",
    "email" : "CoreySchafer@mail.com"
}
```

Suntem familiari cu dicționariile în Python, unde avem perechi de valori de tipul key:value. Dacă reprezentăm anumite date (o persoană în cazul de față) avem un 'key' denumit 'first_name' ce are valoare 'Corey', un 'key' denumit 'last_name' ce are valoare 'Schafer', respectiv 'email' cu valoarea 'CoreySchafer@mail.com '. Dicționarul de mai sus reprezintă date pentru o singură persoană. Prin ce modalitate putem stoca date pentru mai multe persoane?

Sunt mai multe modalități prin care putem face asta, dar modul cel mai bun de a ne gândi la un asemenea tip de stocare a datelor când învățăm pandas este să facem valoarile din cadrul dicțioarului liste

```python
peoples = {
    "first_name" : ["Corey"],
    "last_name" : ["Schafer"],
    "email" : ["CoreySchafer@mail.com"]
}
```

Putem observa că avem un dicționar foarte asemănător cu ce am avut înainte, dar acuma în loc de un simplu string pentru parte de value avem o listă de valori. Momentan, lista are doar câte o singură valoare pentru fiecare 'key', dar din moment ce este o listă putem să adăugăm mai multe valori pentru fiecare key ('first_name', 'last_name' și 'email')

```python
peoples = {
    "first_name" : ["Corey", "Jane", "John"],
    "last_name" : ["Schafer", "Doe", "Doe"],
    "email" : ["CoreySchafer@mail.com", "JaneDoe@mail.com", "JohnDoe@mail.com"]
}
```

În cel de-al 3-lea exemplu pentru fiecare key avem mai multe valori (3 pentru fiecare) în lista respectivă. Prin această metodă, primul element din listă va reprezenta prima persoană, al doilea element a treia persoană etc. Dacă dorim informații despre a doua persoană, trebuie să luăm al doilea element din fiecare listă. Putem să ne imaginăm ca și un tabel acuma, unde partea de 'key' ar reprezenta o coloană, iar partea de 'values' ar reprezenta rândurile din cadrul tabelului.

Cele mai multe definții pentru un data frame în pandas spune că este un tip de date 'two dimensional', ceea ce practic înseamnă rânduri și coloane. Cum spuneam, un 'key' ar reprezenta o coloană într-un data frame. Dacă dorim să extragem toate informații pentru 'email' (key) atunci putem să accesăm acel 'key' al dicționarului. Valorile unui 'key' dintr-un dicționar se accesează utilizând un set de paranteze pătrate în care se introduce numele 'key'-ului pe care dorim să îl accesăm

```python
peoples['email']
```

Acest mod ne ajută să ne imaginăm cum arată un data frame în Python, deoarece sunt foarte asemănătoare cu ceea ce am văzut mai sus, dar cu mult mai multe funcționalități. Putem să creem un data frame din acest dicționar (peoples) și să vedem cum arată. Pentru a crea un data frame dintr-un dicționar se va utiliza metoda `DataFrame()` din 'pandas' la care se trece ca și argument dicționarul respectiv

```python
peoples_df = pd.DataFrame(peoples)
```

```python
peoples_df
```

După ce am creat un data frame din acel dicționar și am afișat ce anume conține acest data frame putem vedea că ne afișează datele din dicționar sub forma unui tabel (cu rânduri și coloane). În partea stângă avem acele numere (0, 1 și 2) care nu au o coloană denumită deasupra, iar acestea reprezintă un index. Nu o să ne axăm acuma pe partea de index, sunt pentru următoarea parte, dar într-un mod foarte simplu reprezintă o valoare unică pentru rândurile din data frame. Nu este neapărat să fie unicat, dar mai multe în secțiunea de index.

Din moment ce avem o idee despre cum să ne imaginăm un data frame, să încercăm acuma să accesăm valorile pentru coloana 'email'. La fel cum am făcut în cazul dicționarului, putem accesa o coloană precum accesăm un 'key' dintr-un dicționar

```python
peoples_df['email']
```

Putem vedea că în urma codului 'peoples_df['email']' obținem toate datele din coloana 'email' a data frame-ului. Am utilizat partea de Python pentru a ne putea imagina mai bine despre ce este un data frame și cum anume funcționează și am utilizat dicționarii deoarece modul de a accesa elemente atât din dicționarii cât și dintr-un data frame sunt identice. Un data frame însă este mult mai mult decât un simplu dincționar.

Putem vedea că ce anume s-a oferit ca și output când am selecatat datele dintr-un data frame nu este la fel precum output-ul pentru datele selectate din dicționar, iar asta este datorat faptului că ceea ce este returnat ca și date dintr-o singură coloană reprezintă un `Series`. Putem verifica asta prin a verifica tipul de date (utilizând 'type()')

```python
type(peoples_df['email'])
```

Output-ul ne spune că este un tip de date de tipul 'Series'. Un tip de date 'Series' este practic tot o listă de date, dar precum un data frame are mult mai multe funcționalități decât o listă. În termeni de specialitate 'Series' este un tip de date 'one dimensional' ceea ce reprezintă un rând de date. Putem să ne gândim la data frame-uri ca fiind rânduri și coloane și la series ca fiind rândurile unei singure coloane. Precum data frame, și series are un index

Putem accesa o singură coloană a unui data frame precum accesăm un 'key' dintr-un dicționar, dar poate sunt anumite persoane care o să utilizeze notația cu puncte (peoples_df.email)

```python
peoples_df.email
```

Putem observa că dacă utilizăm notația cu puncte obține același rezultat precum accesare ca și un 'key'. Modul prin care se accesează este o preferință personală, dar recomand prima variantă, cea cun paranteze drepte (peoples_df['email']). Motivul pentru utilizarea acestei notații este din posibilitatea ca anumite coloane să poartea același nume precum o anumită metodă sau un anumit atribut prezent în cadrul data frame-ului. De exemplu, am aruncat o privire peste atributul '.shape' din data frame ('peoples_df.shape'). Dacă avem o coloană care poartă aceași denumire, atunci utilizând notația cu punct o să accesăm atributul, nu datele ce sunt prezente în cadrul coloanei respective.

Am tot prezizat că un data frame are mai multe funcționalități decât ce am văzut prin utilizarea limbajului standard de Python. Să luăm cazul în care dorim să accesăm mai multe coloane dintr-un data frame. Pentru a accesa mai multe coloane putem utiliza notarea cu paranteze drepte și să îi oferim ca și argument o listă de nume de coloane pe care dorim să le accesăm. 

```python
peoples_df[['last_name', 'email']]
```

Observăm că avem un data frame returnat în acest moment. De menționat că am pasat o listă în cadrul parantezelor drepte, de aceea sunt 2 perechi de paranteze drepte. Nu putem să renunțăm la setul de paranteze din interior deoarece cel mai probabil o să se ridice o eroare (key error) deoarece pandas crede că pasăm ambele string-uri ca și un singur nume de coloană. Acum că avem 2 coloane ce sunt rezultate, ceea ce este în output nu mai poate fi un tip de date 'Series', deoarece un astfel de tip de date reprezintă rândurile dintr-o singură coloană. Când se retunrează mai multe coloane, practic se retunrează un nou data frame (un data frame filtrat doar cu acele coloane)

În situația în care avem un număr mare de coloane în data frame (precum avem 85 în fișierul 'sruvey_results_public.csv') putem să afișăm numele tuturor coloanelor utilizând atributul `.columns`

```python
peoples_df.columns
```

Am văzut cum anume putem obiține o coloană sau mai multe dintr-un data frame, dar cum se procedează pentru un rând? Pentru a returna anumite rânduri dintr-un data frame putem utiliza indexatorii `loc` și `iloc`.

O să începem cu `iloc`. Acest indexator ne permite să accesăm rânduri dintr-un data frame printr-o locație a unui integer (de unde și denumirea 'iloc', integer location). Dacă dorim să returnăm primul rând din data frame urilizâm comanda:

```python
peoples_df.iloc[0]
```

Din moment ce este un indexator o să utilizăm paranteze drepte unde pasăm un index, 0 în cazul de față, care retunrează primul rând. Codul de mai sus retunrează un tip de date 'Series' care conține valorile primului rând de date. În acest caz index-ul pentru elementele returnate reprezintă  numele coloanelor (încă nu am discutat în detalii despre index) pentru a ști ce anume reprezintă acele valori. Când accesăm datele dintr-o coloană o să treacă index-ul precum un integer, dar când accesăm datele dintr-un tabel, ca și index va trece denumirea coloanei din tabel. Astfel știm că pe rândul 1 din data frame pentru coloana 'first_name' avem valoarea 'Corey'. Ar fi destul de confuz să pună index tot un integer deoarece nu prea am ști ce reprezintă acele numere din index.

Precum am selectat coloane multiple, putem selecta și rânduri multiple oferin o listă de valori integer care reprezintă poziția rândurilor din data frame

```python
peoples_df.iloc[[0, 1]]
```

Output-ul de la 'peoples_df.iloc[[0, 1]]' retunrează valorile din primele două rânduri. Din moment ce avem date din mai multe rânduri și mai multe coloane, în acest moment se returnează un data frame.

 Cu acești indexatori ('iloc' și 'loc') putem de asemenea să selectăm și anumite coloane pentru rândurile respective (nu putem selecta doar coloane, ci putem selecta rânduri și coloane specifice). Coloanele ce dorim să le selectăm o să fie trecute ca și valorea a doua pentru indexator, dacă ne gândim la 'iloc' și 'loc' ca ca și cum ar fi funcții, atunci rândurile ce dorim să le selectăm reprezintă primul argument, iar coloanele reprezintă al doilea argument

Cu 'iloc' nu putem specifica nume de coloane deoarece acest indexator folosește integer, prin urmare trebuie să utilizăm tot valori numerice. Numerotarea coloanelor se face de la 0, 'first_name' fiind coloana de pe index-ul 0, 'last_name' de pe index-ul 1, etc. Deci pentru a selecta doar coloana 'email' pentru primele două rânduri trebuie să selectăm și coloana de pe index-ul 2 ('email' este a treia coloană din data frame)

```python
peoples_df.iloc[[0, 1], 2]
```

Acesta este comportamentul indexatorului 'iloc'. Acuma o să aruncăm o privire peste indexatorul `loc`. Utilizând 'iloc' căutam în data frame după localizarea unui index de tipul integer, iar cu indexatorul `iloc` o să căutăm după un label. Când discutăm de 'labels' pentru rânduri, acestea sunt reprezentate de index (momentan nu avem un index customatizat), momentan acest index este un range de valori integer. Acum poate părea la fel precum indexatorul 'iloc', dar o să trecem peste anumite use case-uri utilizând 'loc' cu anumite label-uri și pentru rânduri. Să mai aruncăm o privire peste întregul data frame

```python
peoples_df
```

Valorile ce sunt prezente în partea stângă (0, 1 și 2) reprezintă indexii rânduri, iar dacă dorim primul rând, acesta are un index default de 0

```python
peoples_df.loc[0]
```

Precum 'iloc', și aici putem utiliza o listă pentru a selecta mai multe rânduri.

```python
peoples_df.loc[[0, 1]]
```

De asemenea putem să introducem o a doua valoare în cadrul indexatorului pentru a selecta coloane specifice pentru aceste rânduri. Cu 'iloc' utilizam integer pentru a specifica poziția coloanelor, dar acum utilizând indexatorul 'loc' putem utiliza labels (numele coloanei). Dacă mai dorim pe lângă primele două rânduri să specificăm doar coloane 'email', putem să o trecem ca și un al doilea argument în cadrul indexatorului

```python
peoples_df.loc[[0, 1], 'email']
```

Nu am specificat acest lucru în cadrul indexatorului 'iloc', dar putem de asemenea să specificăm o listă pentru partea de coloane de asemenea, dacă dorim să selectăm mai mult de o coloană pentru anumite rânduri. Dacă dorim atât coloana 'last_name' și 'email' pentru primele 2 rânduri, în loc de a specifica un string pentru cea de-a doua valoare, putem specifica o listă de string-uri în care fiecare string să reprezinte o coloană din data frame

```python
peoples_df.loc[[0, 1], ['email', 'last_name']]
```

În cadrul data frame-ului ordinea în care sunt trecute coloanele este 'first_name', 'last_name' și 'email', dar din moment ce în indexatorul 'loc' s-a specificat prima dată coloane 'email' înainte de coloane 'last_name', prima dată se afișează coloana 'email', nu ține cont de ordinea în care sunt trecute în data frame, ține cont de ordinea în care sunt trecute în lista de coloane din 'loc'

Din moment ce am văzut cum funcționează aceste procedee de a selecta anumite rânduri sau coloane cu un set mic de date, o să trecem din nou la data frame-ul de la studiul celor de la stackoverflow ('survey_results_public.csv') în care avem un număr mult mai mare de coloane și de rânduri. Acum să ne întoarcem din nou la partea de selectare a datelor. Până acuma din data frame-ul nostru am selectat toate datele dintr-o anumită coloană ('df['Hobbyist']'), acum să selectăm atât un rând specific cât și o coloană specifică. O să selectăm primul rând și de asemenea, tot coloana 'Hobbyist'. Pentru a putea selecta atât un rând cât și o coloană se poate utiliza unul dintre indexatorii 'loc' sau 'iloc'. O să utilizăm indexatorul 'loc' deoarece ne permite să selectăm anumite rânduri sau coloane prin specificarea de labels, iar pentru coloana ce dorim să o selectăm o să ne folosim de label (numele coloanei)

Din moment ce utilizăm un index default și nu unul customatizat, curentul label pentru rânduri reprezintă doar un set de valori întregi de la 0 până la aproape 89.000 (câte răspunsuri sunt în data frame)

Pentru indexatori putem să trecem ca și argumente o listă de valori pentru a selecta mai multe rânduri sau coloane dintr-un data frame. Dacă dorim să extragem valorile din primele 3 rânduri pentru coloana 'Hobbyist' atunci putem să trecem o listă de valori de tip integer, și anume [0, 1, 2], deoarece acestea reprezintă primele 3 rânduri din data frame

O funcționalitate pe care nu am văzut-o până acuma este că putem utiliza conceptul de 'slicing' pentru a selecta multiple rânduri sau coloane. Diferența pentru slicing la un data frame este faptul că ultima valoare din acel interval de slicing este inclusă în selectare, cel puțin cu indexatorul 'loc'. Dacă dorim la fel să selectăm primele 3 rânduri, în loc de valoril [0, 1, 2] putem trece ca și un slicing aceste valori 

```python
df.loc[[0:2], 'Hobbyist']
```

Oops! De fiecare dată când utilizăm conceptul de slicing pentru un data frame nu trebuie să îl trecem între paranteze drepte, precum o listă 

```python
df.loc[0:2, 'Hobbyist']
```

Utilizând conceptul de slicing am primit același rezultat ca și mai sus, în care am setat manual numărul rândurilor ce dorim să le selectăm. Putem să utilizîm conceptul de slicing și pentru partea de coloane. Numele coloanelor pot fi trecute ca și valori pentru partea de slicing. Acesta este motivul pentru care ulrima valoare este inclusă în partea de slicing într-un data frame. Era un pic frustrant să dorim toatele coloane de la 'Hobbyist' până la 'Employment', dar ultima coloană din slicing să nu fi fost inclusă. Așa trebuia să ne uităm din nou în numele coloanelor din data frame și să vedem ce coloană vine după 'Employment' (Country) și să precizăm acea coloană în partea de slicing

```python
df.loc[0:2, 'Hobbyist':'Employment']
```
